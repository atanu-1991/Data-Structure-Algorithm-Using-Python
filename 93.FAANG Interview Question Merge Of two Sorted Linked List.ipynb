```python
## Given the user, two sorted Linked List. The expected output is to give a fully sorted Linked List in a merge order.
## Time Complexity: 
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    ## insert at the very End of the linked list
    def insertAtEnd(self, new_data):
        ## creation of the new node
        new_node = Node(new_data)
        
        ## Linked List is empty
        if self.head is None:
            self.head = new_node
            return
        
        ## insertion at the end
        temp = self.head
        while temp.next:
            temp = temp.next
            
        temp.next = new_node
    
    
    ## print the Linked List
    def printList(self):
        temp = self.head
        while temp:
            print(str(temp.data)+" ",end=" ")
            temp = temp.next

## method definition to merge Lists
def mergeList(head1, head2):
    temp = None
    ## LList1 is empty
    if head1 is None:
        return head2
    
    ## LList2is empty
    if head2 is None:
        return head1
    
    if head1.data < head2.data:
        temp = head1
        ## recursive function calling is happened
        temp.next = mergeList(head1.next, head2)
    else:
        temp = head2
        ## recursive function calling is happened
        temp.next = mergeList(head1, head2.next)
        
    return temp

            
## Driver Code
llist1 = LinkedList()
## function calling
llist1.insertAtEnd(10)
llist1.insertAtEnd(20)
llist1.insertAtEnd(40)
llist1.insertAtEnd(70)
llist1.insertAtEnd(100)
print("First linked List:")
llist1.printList()
print()

llist2 = LinkedList()
## function calling
llist2.insertAtEnd(8)
llist2.insertAtEnd(12)
llist2.insertAtEnd(17)
llist2.insertAtEnd(27)
llist2.insertAtEnd(49)
print("Second linked List:")
llist2.printList()
print()

llist3 = LinkedList()
llist3.head = mergeList(llist1.head, llist2.head)
print("Merge Lists")
llist3.printList()
```
